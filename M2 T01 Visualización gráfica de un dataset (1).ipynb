```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#cargamos el data set modificado con las nuevas variables creadas en la tarea anterior:
df=pd.read_csv(r'C:\Users\wcc30\Documents\TascaPython.csv', index_col = 0)

#info
df.info()
```

```python
#Visualizamos el dataframe
df
```

#### Una variable categórica (UniqueCarrier)

```python
plt.figure(figsize=(12,8))
plot = sns.countplot(y="UniqueCarrier", data=df,order = df.UniqueCarrier.value_counts().index,palette = 'mako')
plt.title('Número de vuelos realizados según Unique Carrier Code en el año 2008')
plt.xlabel('Número de vuelos')
plt.ylabel('Unique Carrier Code')

#Guardamos la visualizacion
plt.savefig('UniqueCarrier.png')
```

#### Una variable numérica (ArrDelay)

```python
plt.figure(figsize=(12,8))
df['ArrDelay'].plot.hist()
plt.title('Frecuencia de ArrDelay')
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
df['ArrDelay'].describe()
```

Podemos ver que debido a la alta diferencia ente el minimo y maximo valor, la primera visualización de la distribución no es muy clara que digamos, por ello decidimos hacer una escala logaritmica para poder obtener una mejor visualización sobre su distribución y a la vez nos permitirá ver si tenemos valores outliers, en caso exista.

```python
plt.figure(figsize=(12,8))
df['ArrDelay'].plot.hist()
plt.yscale("log")
plt.title('Frecuencia de ArrDelay')

#Guardamos la visualizacion:
plt.savefig('ArrDelay.png')
```

Como sospechabamos, encontramos valores outliers y lo podemos visualizar en la barra de frecuencia separada de las demás. Además concluimos que los datos de la variable "ArrDelay" tiene una distribución asimétrica positiva.

#### Una variable numérica y una categórica (ArrDelay y UniqueCarrier)

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="UniqueCarrier", y="ArrDelay", data=df)
```

Realizamos grafico de cajas para cada item de la variable UniqueCarrier usando los valores numéricos de ArrDelay pero debido a sus demasiados outliers la visualización es pésima, por lo cual realizaremos un filtro a los datos tomando en cuenta solo los valores menores a 100 de la variable ArrDelay.

```python
plt.figure(figsize=(12,8))
filtro = df['ArrDelay'] < 100
sns.boxplot(x="UniqueCarrier", y="ArrDelay", data=df[filtro])

#Guardamos la visualización
plt.savefig('ArrDelayvsUniqueCarrier.png')
```

Podemos ver una mejor visualización de los datos , pudiendo identificar valores por ejemplo: En el item XE podemos ver que su mediana esta cerca a los 40.

Otro gráfico que nos podría servir seria un grafico de barras tomando en cuenta el promedio de ArrDelay segun Unique Carrier Code.

```python
plt.figure(figsize=(12,8))
sns.barplot(x = 'UniqueCarrier',y = 'ArrDelay', data = df, estimator = np.mean, ci = 0, palette = "hls")
plt.title('Mean Arrive Delay by Unique carrier', size = 16)  
plt.xlabel("Unique Carrier")
plt.ylabel("Arrive Delay")
plt.show()

#Guardamos la visualización
plt.savefig('MEANArrDelayvsUniqueCarrier.png')
```

#### Dos variables numéricas (ArrDelay y DepDelay)

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="ArrDelay", y="DepDelay")
plt.title('Dispersión ArrDelay vs DepDelay', size = 16) 

#Guardamos la visualización
plt.savefig('DispersiónArrDelayvsDepDelay.png')
```

Mediante el grafico de dispersión aplicada a las dos variables numéricas podemos visualizar que tienen una correlacion lineal positiva y ademas valores outliers

#### Tres variables (ArrDelay, DepDelay y UniqueCarrier)

Podemos utilizar la grafica anterior de dispersión , agregandole la variable categórica para poder diferenciar mediante 
los colores los distintos puntos según Unique Carrier Code.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="ArrDelay", y="DepDelay", hue = "UniqueCarrier")
plt.title('Dispersión ArrDelay vs DepDelay by UniqueCarrier', size = 16) 

#Guardamos la visualización
plt.savefig('DispersiónArrDelayvsDepDelaybyCarrier.png')
```

```python
mean_delay = df.groupby('UniqueCarrier')['ArrDelay', 'DepDelay'].mean()

plt.figure(figsize=(12, 8))
plt.bar(mean_delay.index, mean_delay.ArrDelay, width = 0.5, color = 'green', label = 'Arr Delay')
plt.bar(mean_delay.index, mean_delay.DepDelay, width = 0.5, align = 'edge', color = 'blue',label = 'Dep Delay')
plt.title('Mean ArrDelay vs DepDelay by UniqueCarrier', size = 16) 
plt.legend()
plt.show()

#Guardamos la visualización
plt.savefig('ArrDelayvsDepDelaybyCarrierMEAN.png')
```

#### Más de tres variables (ArrDelay, DepDelay, AirTime y UniqueCarrier)

```python
f, ax = plt.subplots(figsize=(12, 12))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="ArrDelay", y="DepDelay",
                hue="UniqueCarrier", size="AirTime",
                sizes=(1, 12), linewidth=0,
                data=df, ax=ax)
plt.show()

#Guardamos la visualización
plt.savefig('4Variables.png')
```

Hemos vuelto a usar el grafico de dispersión , pero en este caso usando la variable "AirTime"como el tamaño de las burbujas según su valor, mientras mayor sea el valor de "Air time" más grande sera la burbuja
